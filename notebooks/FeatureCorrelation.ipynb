```python
corr_matrix = dfother.corr().abs()

upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(np.bool))

to_drop = [column for column in upper.columns if any(upper[column] > 0.5)]
to_drop
```

```python
dfother.drop(to_drop, axis=1)
```

```python
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from scipy.sparse import hstack
```

```python
btr,bte,gtr,gte = train_test_split(df.loc[:, df.columns != 'Gender'], df.Gender, test_size=0.25, shuffle=True, stratify=df.Gender)
```

```python
cv = CountVectorizer()
cv.fit(btr.Blog.astype(str))
btrBlog = cv.transform(btr.Blog.astype(str))
bteBlog = cv.transform(bte.Blog.astype(str))
```

```python
clf = MLPClassifier(hidden_layer_sizes=(25,25), solver='adam', activation='relu', early_stopping=True)
```

```python
clf.fit(hstack((btrBlog, dfother), format='csr'), gtr)
```

```python
clf.score(bte, gte)
```
