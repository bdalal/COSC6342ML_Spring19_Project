```python
import numpy as np
import pandas as pd
import os
```

```python
path = '../../ProjectDatasets/Gender/koppel_blogs/blogs/'
files = os.listdir(path)
```

```python
blogs = []
ctr = 1

for file in files:
    print('Working on file', ctr, file, 'of 19320')
    gender = file.split('.')[1]
    if gender == 'male':
        gender = 'M'
    else:
        gender = 'F'
    with open(path + file, 'r', encoding='iso-8859-15') as blogfile:
        for possible_blog in (''.join(blogfile.readlines())).split('</post>'):
            if '<post>' in possible_blog:
                blog = possible_blog[possible_blog.find('<post>') + 6:].strip()
                blog_entry = (blog, gender)
                blogs.append(blog_entry)
    ctr += 1
```

```python
df = pd.DataFrame(blogs, columns=['Blog', 'Gender'])
```

```python
df = df.sample(frac=1).reset_index(drop=True)
```
