```python
import pandas as pd
import numpy as np
from pickle import load, dump
from flair.data import Sentence
from flair.models import SequenceTagger
```

```python
data = pd.read_csv('../data/blog-gender-dataset_csv.csv')
```

```python
tagger = SequenceTagger.load('pos-fast')
```

```python
blog_tags = []
ctr = 1
for blog in data.Blog.values:
    print("Working for Blog", ctr)
    sentence = Sentence(str(blog))
    tagger.predict(sentence)
    tag_list = []
    for token in sentence.tokens:
        tag_list.append(token.tags['pos'].value)
    blog_tags.append(tag_list)
    ctr += 1
```

```python
with open('../data/POS_Tags.pkl', 'wb') as pkldump:
    dump(blog_tags, pkldump)
```

```python
print('Blog = ', data.Blog.values[0])
print('\nPOS = ', ' '.join(blog_tags[0]))
```

```python
from collections import Counter

tagset = {'CC', 'CD', 'DT', 'EX', 'FW', 'IN', 'JJ', 'JJR', 'JJS', 'LS', 'MD', 'NN', 'NNS', 'NNP', 
          'NNPS', 'PDT', 'POS', 'PRP', 'PRP$', 'RB', 'RBR', 'RBS', 'RP', 'SYM', 'TO', 'UH', 'VB', 'VBD', 
          'VBG', 'VBN', 'VBP', 'VBZ', 'WDT', 'WP', 'WP$', 'WRB'}
tag_counter = Counter(dict.fromkeys(tagset, 0))
```

```python
for tags in blog_tags:
    for tag in tags:
        if tag in tagset:
            tag_counter[tag] += 1
```

```python
with open('../data/Tag_Counts.pkl', 'wb') as pkldump:
    dump(tag_counter, pkldump)
```

```python
from itertools import tee, islice
def ngrams(lst, n):
    tlst = lst
    while True:
        a, b = tee(tlst)
        l = tuple(islice(a, n))
        if len(l) == n:
            yield l
            next(b)
            tlst = b
        else:
            break
```

```python
ngram_counter = Counter(dict.fromkeys(tagset, 0))
ctr = 1
for tags in blog_tags:
    print('Working for blog', ctr)
    for i in range(1, 8):
        ngram_counter += Counter(ngrams(tags, i))
    ctr += 1
```

```python
with open('../data/NGramCounts.pkl', 'wb') as pkldump:
    dump(ngram_counter, pkldump)
```
